```python
import pandas as pd
# activity = pd.read_csv('trainer_activity.csv')
activity = pd.read_csv('activity.csv')
```

```python
import matplotlib.pyplot as plt
plt.plot(activity['watts'], label='Raw Watts')
for duration in [30, 60, 120, 300, 360, 600]:
    activity[f'{duration}s_avg_watts'] = activity['watts'].rolling(duration).mean()
    plt.plot(activity[f'{duration}s_avg_watts'], label=f'{duration}s avg Watts')
plt.legend()
```
